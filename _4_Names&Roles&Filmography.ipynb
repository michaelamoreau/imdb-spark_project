```
#Daniel RadCliffe
print("Filmography:")
print("=======================================================")
import imdb
moviesDB = imdb.IMDb()
ActorId = "0705356"
person = moviesDB.get_person(ActorId)
print (person)
actor_work = moviesDB.get_person_filmography(ActorId)
for movie in actor_work['data']['filmography']['actor'] :
    try:
        print (movie.movieID, movie['title'], movie.currentRole, movie['year'])
    except KeyError:
       
        print (movie.movieID, movie['title'], movie.currentRole)
print("=======================================================")
print("Roles:")
print("=======================================================")
import requests
from bs4 import BeautifulSoup
url = 'https://www.imdb.com/name/nm0705356/'
soup = BeautifulSoup(requests.get(url).content, 'html.parser')

seen = set()
for row in soup.select('#filmo-head-actor + div .filmo-row > br'):
    role = row.find_next(text=True).strip()
    if not role in seen:
        seen.add(role)
        print(role)
print("=======================================================")

```
